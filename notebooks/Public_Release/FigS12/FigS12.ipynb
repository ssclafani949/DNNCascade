```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
mpl.rcParams['text.usetex'] = True
mpl.rc('font', size=18) 
```

```python
E2_sindec = np.genfromtxt('E2_sindec.csv', delimiter=',')
Sensitvity_E2 = np.genfromtxt('Sensitvity_E2.csv', delimiter=',')

E2_sindec_dp = np.genfromtxt('E2_sindec_dp.csv', delimiter=',')
DP_4sigma_E2 = np.genfromtxt('DP_4sigma_E2.csv', delimiter=',')
```

```python
E3_sindec = np.genfromtxt('E3_sindec.csv', delimiter=',')
Sensitvity_E3 = np.genfromtxt('Sensitvity_E3.csv', delimiter=',')

E3_sindec_dp = np.genfromtxt('E3_sindec_dp.csv', delimiter=',')
DP_4sigma_E3 = np.genfromtxt('DP_4sigma_E3.csv', delimiter=',')
```

```python
source_list_ULS = np.genfromtxt('source_list_ULS.csv', delimiter=',', skip_header=1)
```

```python
UL_decs = source_list_ULS[:,3]
UL_E2 = source_list_ULS[:,1]
UL_E3 = source_list_ULS[:,2]
```

```python
colors = ['#0B3D53', '#FF7F0E',  '#5C9FC9']
```

```python
fig, (ax1, ax2) = plt.subplots (1, 2, figsize=(12,6)) 
 
ax1.plot(E2_sindec, Sensitvity_E2, label='This Work', color=colors[0], lw=3)
ax1.plot(E2_sindec, DP_4sigma_E2, label='This Work 4$\sigma$', color=colors[0], ls='--', lw=2)

ax1.scatter(UL_decs, UL_E2, marker='v', color='dimgrey', 
            s=25, label='Upper Limit 90\% C.L.', zorder=10,  rasterized=True)


ax2.plot(E3_sindec, Sensitvity_E3, label='This Work', color=colors[0], lw=3)
ax2.plot(E3_sindec, DP_4sigma_E3, label='This Work 4$\sigma$', color=colors[0], ls='--', lw=2)
ax2.scatter(UL_decs, UL_E3, marker='v', color='dimgrey', 
            s=25, label='Upper Limit 90\% C.L.', zorder=10,  rasterized=True)


ax1.set_xlabel (r'$\sin(\delta)$')
ax1.set_ylabel (
    r'$E^2 \cdot dN/dE$ at 100 TeV [$\textsf{TeV}\,\textsf{s}^{-1} \textsf{cm}^{-2}]$')
ax1.set_xlim (-1, 1)
ax1.set_ylim (.5e-13, 5e-10)
ax1.legend (ncol=1, fontsize=14)
ax1.set_title(r'dN/dE $\propto$ E$^{-2}$')

ax2.set_xlabel (r'$\sin(\delta)$')
ax2.set_xlim (-1, 1)
ax2.set_ylim (.5e-13, 5e-10)


ax2.set_title(r'dN/dE $\propto$ E$^{-3}$')
ax1.semilogy()
ax2.semilogy()
plt.tight_layout()
```
