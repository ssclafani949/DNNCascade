```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.rc('font', size=16)

plt.rcParams['axes.facecolor']='w'
plt.rcParams['savefig.facecolor']='w'
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}']
colors = ['#0B3D53', '#FF7F0E',  '#5C9FC9']
```

```python
effa_cascades_bins = np.loadtxt('effa_cascades_bins.csv')
effa_cascades_values = np.loadtxt('effa_cascades_values.csv')

effa_tracks_bins = np.loadtxt('effa_tracks_bins.csv')
effa_tracks_values = np.loadtxt('effa_tracks_values.csv')

effa_dnnc_bins = np.loadtxt('effa_this_work_bins.csv')
effa_dnnc_values = np.loadtxt('effa_this_work_values.csv')
```

```python
nevents_cascades_bins = np.loadtxt('nevents_cascades_bins.csv')
nevents_cascades_values = np.loadtxt('nevents_cascades_values.csv')

nevents_tracks_bins = np.loadtxt('nevents_tracks_bins.csv')
nevents_tracks_values = np.loadtxt('nevents_tracks_values.csv')

nevents_dnnc_bins = np.loadtxt('nevents_this_work_bins.csv')
nevents_dnnc_values = np.loadtxt('nevents_this_work_values.csv')
```

```python
plt.rc('font', size=18)
label = 'ps_tracks_cascades'
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,6))

#Effective Area Plots
ax1.step(effa_tracks_bins[1:], effa_tracks_values, label=r'Tracks [\textsf{IceCube (2020)}]', 
          color = 'grey', lw=3, ls='-', zorder=-1,)
ax1.step(effa_cascades_bins[1:], effa_cascades_values, label=r' Cascades [\textsf{IceCube (2019)}]', 
          color = colors[2], lw=3, ls='-', zorder=-1)

ax1.step(effa_dnnc_bins[1:], effa_dnnc_values, label=r'\textsf{This Work}', color = colors[0], lw=4,ls='-', )

#N events plots
ax2.step(nevents_dnnc_bins[1:], nevents_dnnc_values, 
         lw=4,  color = colors[0])
ax2.step(nevents_cascades_bins[1:], nevents_cascades_values,
         lw=3,   color = colors[2])
ax2.step(nevents_tracks_bins[1:], nevents_tracks_values,
         lw=3,  color = 'grey')

ax2.loglog()
ax2.set_ylabel(r'N$_\textsf{Astro}$ [year$^{-1}$] (-90$^\circ$ $<$ $\delta$ $<$ -5$^\circ$)')

ax2.set_ylim(1e-2,20)
ax2.set_yticks([1e-2, 1e-1, 1,10])
#fig.suptitle('-90$^\circ$ $<$ $\delta$ $<$ -5$^\circ$')

handles, labels = ax1.get_legend_handles_labels()
ax1.legend(handles[::-1], labels[::-1])#, loc='lower right')

ax1.set_ylabel('A$_\mathsf{Eff}$ [m$^2$] (-90$^\circ$ $<$ $\delta$ $<$ -5$^\circ$)')
ax2.set_xlabel(r'E$_\nu$ [GeV]')
ax1.set_xlabel(r'E$_\nu$ [GeV]')


ax1.loglog()
ax1.set_xlim(700, 1e6)
ax2.set_xlim(700, 1e6)

ax1.set_ylim(1e-5,2e2)
plt.tight_layout()
```
