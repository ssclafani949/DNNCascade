```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['text.usetex'] = True
```

```python
astro_tot_bins = np.genfromtxt('astro_tot_bins.csv', delimiter=',')
astro_tot_values = np.genfromtxt('astro_tot_values.csv', delimiter=',')

astro_nue_bins = np.genfromtxt('astro_nue_bins.csv', delimiter=',')
astro_nue_values = np.genfromtxt('astro_nue_values.csv', delimiter=',')
astro_numu_bins = np.genfromtxt('astro_numu_bins.csv', delimiter=',')
astro_numu_values = np.genfromtxt('astro_numu_values.csv', delimiter=',')
astro_nutau_bins = np.genfromtxt('astro_nutau_bins.csv', delimiter=',')
astro_nutau_values = np.genfromtxt('astro_nutau_values.csv', delimiter=',')

atmo_tot_bins = np.genfromtxt('atmo_tot_bins.csv', delimiter=',')
atmo_tot_values = np.genfromtxt('atmo_tot_values.csv', delimiter=',')

atmo_nue_bins = np.genfromtxt('atmo_nue_bins.csv', delimiter=',')
atmo_nue_values = np.genfromtxt('atmo_nue_values.csv', delimiter=',')
atmo_numu_bins = np.genfromtxt('atmo_numu_bins.csv', delimiter=',')
atmo_numu_values = np.genfromtxt('atmo_numu_values.csv', delimiter=',')
atmo_nutau_bins = np.genfromtxt('atmo_nutau_bins.csv', delimiter=',')
atmo_nutau_values = np.genfromtxt('atmo_nutau_values.csv', delimiter=',')
```

```python
cmap = plt.get_cmap('RdBu')
```

```python
plt.rc('font', size=15)
fig, ax = plt.subplots(figsize=(7,6))
fig.set_facecolor('w')

plt.step(astro_tot_bins[1:], astro_tot_values,  label=r'Total Astrophysical', color=cmap(.99), lw=5, ls = '-', )

plt.step(astro_nue_bins[1:], astro_nue_values,  label = r'Astro $\nu_e$', color=cmap(.9), lw=4, ls = (0, (5, 1)) , alpha=0.85,)
plt.step(astro_numu_bins[1:], astro_numu_values,  label = r'Astro $\nu_\mu$', color=cmap(.74), lw=4 ,ls = ':',  alpha=0.85,)
plt.step(astro_nutau_bins[1:], astro_nutau_values,  label = r'Astro $\nu_\tau$', color=cmap(.6),lw=4,  ls=(0, (3, 1, 1, 1)),  alpha=0.85, )


plt.step(atmo_tot_bins[1:], atmo_tot_values,label=r'Total Atmospheric', color=cmap(0), lw=5, ls = '-', alpha=1)

plt.step(atmo_nue_bins[1:], atmo_nue_values,  color=cmap(.1), label = r'Atmo $\nu_e$',lw=4, ls = (0, (5, 1)), alpha=0.85)
plt.step(atmo_numu_bins[1:], atmo_numu_values,  color=cmap(.25),label = r'Atmo $\nu_\mu$',  lw=4, ls =':', alpha=0.85)
plt.step(atmo_nutau_bins[1:], atmo_nutau_values,  color=cmap(.35), label= r'Atmo $\nu_\tau$', lw=4,ls=(0, (3, 1, 1, 1)), alpha=0.85)


plt.loglog()

ax.set_xlim(600,1e6)
ax.set_ylim(1,5e6)
ax.set_xlabel(r'E$_\nu$ [GeV]')
ax.set_ylabel('Number of Events')
ax.legend(ncol=2)
plt.tight_layout()
```
