```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['text.usetex'] = True
mpl.rc('font', size=14) 
```

```python
data_bins = np.genfromtxt('sindec_plot/data_bins.csv', delimiter=',')
data_values = np.genfromtxt('sindec_plot/data_values.csv', delimiter=',')
data_errors = np.genfromtxt('sindec_plot/data_errors.csv', delimiter=',')

astro_nu_bins = np.genfromtxt('sindec_plot/astro_nu_bins.csv', delimiter=',')
astro_nu_values = np.genfromtxt('sindec_plot/astro_nu_values.csv', delimiter=',')

atmo_nu_bins = np.genfromtxt('sindec_plot/atmo_nu_bins.csv', delimiter=',')
atmo_nu_values = np.genfromtxt('sindec_plot/atmo_nu_values.csv', delimiter=',')

mu_bins = np.genfromtxt('sindec_plot/mu_bins.csv', delimiter=',')
mu_values = np.genfromtxt('sindec_plot/mu_values.csv', delimiter=',')

sim_tot_bins = np.genfromtxt('sindec_plot/sim_tot_bins.csv', delimiter=',')
sim_tot_values = np.genfromtxt('sindec_plot/sim_tot_values.csv', delimiter=',')
sim_tot_errors = np.genfromtxt('sindec_plot/sim_tot_errors.csv', delimiter=',')
```

```python
colors = ['#FF7F0E', '#0B3D53', '#5C9FC9', ]
```

```python
bins = 35
range = (-1,1)
fig, (ax1, ax2) = plt.subplots(2,1, figsize=(8.25,6), gridspec_kw = {'height_ratios':[3, 1]}, sharex=True)


ax1.errorbar((data_bins[:-1] + data_bins[1:])/2, data_values, c= 'k', drawstyle='steps-post',yerr=data_errors,xerr=np.diff(data_bins)/2,
             linestyle='none', label='Data')


ax1.step(mu_bins, np.r_[0,atmo_nu_values + mu_values], color= colors[1])
ax1.step(mu_bins, np.r_[0,mu_values], color= colors[2])
ax1.fill_between(mu_bins, np.r_[0,mu_values], color=colors[2], alpha = 0.3, step='pre',  label=r'Atmo $\mu$')
ax1.fill_between(mu_bins, np.r_[0,mu_values], np.r_[0,mu_values+atmo_nu_values], color=colors[1], 
                 alpha = 0.3, step='pre', label=r'Atmo $\nu$')
ax1.fill_between(mu_bins, np.r_[0,mu_values+atmo_nu_values], np.r_[0,mu_values+atmo_nu_values+astro_nu_values],
                 color=colors[0], alpha = 0.3, step='pre', label=r'Astro $\nu$')
ax1.step(sim_tot_bins, np.r_[0,sim_tot_values], c='darkred', label=r'Total $\nu$ and $\mu$ simulation')


data_mc_ratio = data_values/sim_tot_values
errs = (data_values + data_errors) 

errors =   (data_values / sim_tot_values) *np.sqrt(
                        (data_errors / data_values)**2 + (sim_tot_errors / sim_tot_values)**2)

ax2.errorbar((data_bins[:-1] + data_bins[1:])/2, data_mc_ratio, 
             c= 'k', drawstyle='steps-post',yerr= errors, xerr=np.diff(data_bins)/2,
             linestyle='none', label='Data')

ax2.set_xlabel(r'sin($\delta$)')
ax1.set_ylim(0,6000)
ax1.set_ylim(0)
ax1.set_xlim(0-1,1)
ax2.set_ylim(.5, 1.5)
ax2.set_ylabel('Total Data/MC')
ax2.set_yticks([0.5,1.0,1.5])
ax2.axhline(1.0, ls=':', c='k' )
ax1.grid()
ax1.legend(ncol=3)
ax2.grid()
```
