```python
def effa_plot(ax, ana, dec_deg_min = -90, dec_deg_max = 90, label = 'MISC', color = 'k', ls = '-'):
    a = ana.anas[-1]
    mask = (a.sig.dec < np.radians(dec_deg_max)) & (a.sig.dec > np.radians(dec_deg_min))

    dlogE=.05 #.1
    solid_angle=2*np.pi*(np.sin(np.radians(dec_deg_max))-np.sin(np.radians(dec_deg_min)))
    area= 1/ (1e4*np.log(10)) * (a.sig.oneweight[mask] / ( a.sig.true_energy[mask] *  solid_angle *  dlogE))
    if 'GRECO' in label:
        bins = np.arange(.5,6,.1)
    else:
        bins=np.arange(2,7.01,.1)
    h = hl.hist(np.log10(a.sig.true_energy[mask]), weights=area, bins=bins);
    hl.plot1d(h, histtype='step', linewidth=2, color=color, label=label, ls = ls)
    ax.semilogy()

    ax.grid()
    ax.legend(loc='lower right')
    ax.set_ylabel('$A_\mathsf{Eff}$($m^2$)')
    ax.set_xlabel('log$_{10}$(E) [GeV]')
    if 'GRECO' in label:
        ax.set_xlim(.5,6)
        ax.set_ylim((1e-8,1e2))
    else:
        ax.set_xlim((2,7))
        ax.set_ylim((1e-4,1e3))
    plt.tight_layout()
    return h
```

```python
def event_rate(ana):
    a = ana.anas[-1]
    sig = a.sig
    event_rate = len(a.data.ra) / a.bg_livetime
    print('Event Rate {:.3} Hz'.format(event_rate))
    print('{:3f} Events / year'.format(len(a.data.ra) / a.bg_livetime * 86400 * 365))
    
    mask = a.data.dec > np.radians(-5)
    
    event_rate = len(a.data.ra[mask]) / a.bg_livetime
    print('Northern Event Rate {:.3} Hz'.format(event_rate))
    print('{:3f} Events / year'.format(len(a.data.ra[mask]) / a.bg_livetime * 86400 * 365))

    mask = a.data.dec < np.radians(-5)

    event_rate = len(a.data.ra[mask]) / a.bg_livetime
    print('Southern Event Rate {:.3} Hz'.format(event_rate))
    print('{:3f} Events / year'.format(len(a.data.ra[mask]) / a.bg_livetime * 86400 * 365))
```

```python
def angerr(ana, ax, color, label):
    a = ana.anas[-1]
    dpsi = astro.angular_distance(a.sig.dec , a.sig.ra, a.sig.true_dec, a.sig.true_ra)
   
    #plt.rc ('font', size=16)
    #fig, ax = plt.subplots(figsize=(8,7))
    if 'GRECO' in label:
        akw = dict (bins=(15,10**4), range=((50, 1e3), (0, 180)), log=(True,False))

    else:
        akw = dict (bins=(15,10**4), range=((400, 2e6), (0, 180)), log=(True,False))

    weight_E200 = 1e-18 * a.sig.oneweight * (a.sig.true_energy/1e5) ** -2.
    
    hnn = hl.hist_slide ((5,1), (a.sig.true_energy, np.degrees (dpsi)), weight_E200, **akw)
    hnn = hnn.normalize()

    hnn2_baseline = hnn2 = hnn.contain (1, .2)
    hnn5_baseline = hnn5 = hnn.contain (1, .5)
    hnn8_baseline = hnn8 = hnn.contain (1, .8)

    hl.fill_between (ax, 0, hnn5, color=color, alpha=.33, drawstyle='default')
    hl.fill_between (ax, 0, hnn2, color=color, alpha=.33, drawstyle='default')
    hl.fill_between (ax, 0, hnn8, color=color, alpha=.33, drawstyle='default')

    hl.plot1d (ax, hnn2, color=color, lw=2,ls='--', drawstyle='default')
    hl.plot1d (ax, hnn5, color=color, lw=3, drawstyle='default')
    hl.plot1d (ax, hnn8, color=color, lw=2, ls =':', drawstyle='default')

    ax.plot (np.nan, np.nan, color=color, lw=2, alpha=1 , ls = ':', label=r'80\% {}'.format(label))
    ax.plot (np.nan, np.nan, color=color, lw=2, alpha=1 , ls = '-', label=r'50\% {}'.format(label))
    ax.plot (np.nan, np.nan, color=color, lw=2, alpha=1 , ls = '--', label=r'20\% {}'.format(label))

    plt.legend(ncol =1 )
    ax.set_xlabel (r'$E_\textsf{true}$ [GeV]')
    ax.set_ylabel (r'$\Delta\Psi[\text{true,reco}]~[^\circ]$')
    if 'GRECO' in label:
        ax.set_xlim (50, 1e4)
        ax.set_ylim(0,90)
    else:
        ax.set_xlim (500, 1e6)
        ax.set_ylim (0)
    ax.semilogx()
    plt.tight_layout()
    return hnn5
```

```python
fig, ax = plt.subplots(figsize=(8,6))
a = ana.anas[-1]
decs = np.arange(-80,81,5)
livetimet = a.livetime / (86400  * 365)
areast = []
emin = 5e3
emax = 225e3
for dec in decs:
    dec_deg_max = dec + 5
    dec_deg_min = dec
    mask = (a.sig.dec < np.radians(dec + 10)) & (a.sig.dec > np.radians(dec)) & (a.sig.true_energy > emin) & (a.sig.true_energy < emax)
    dlogE= .1 #np.log10(225)- np.log10(5)
    solid_angle=2*np.pi*(np.sin(np.radians(dec_deg_max))-np.sin(np.radians(dec_deg_min)))
    
    area= 1/ (1e4*np.log(10)) * (a.sig.oneweight[mask] / ( a.sig.true_energy[mask] *  solid_angle *  dlogE))
    
    bins=np.arange(3,6.01, dlogE)
    
    h = hl.hist(np.log10(a.sig.true_energy[mask]), weights=area, bins=bins);
    areast.append( h.integrate().values)

plt.plot(decs, np.array(areast) * livetimet, label = 'Tracks')
ax.set_xticks(np.arange(-90,91,15))
ax.grid()
ax.legend(loc='lower right')
ax.set_ylim(0)
ax.set_ylabel(r'Exposure (m$^2$ * yr)')
ax.set_xlabel('Declination (Degrees)')
plt.tight_layout()
#plt.savefig('/home/ssclafani/public_html/DNNC/plots/exposure_m2.png')
```
